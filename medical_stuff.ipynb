```python
dfa_train = pd.read_csv('./train_A_derma.csv')

dfa_trainres = dfa_train['Fake/Real']
# Drop the column from the original DataFrame
dfa_train = dfa_train.drop(['Fake/Real'], axis=1)
# Convert 'fake' to 0 and 'real' to 1
dfa_trainres = dfa_trainres.map({'fake': 0, 'real': 1})

#dfa_train['NaN Size'] = dfa_train[['Mid size', 'Small size', 'Large size']].isnull().sum(axis=1)
dfa_train['NaN SSize'] = dfa_train[[ 'Small size']].isnull().sum(axis=1)
dfa_train['NaN MSize'] = dfa_train[['Mid size']].isnull().sum(axis=1)
dfa_train['NaN LSize'] = dfa_train[['Large size']].isnull().sum(axis=1)
dfa_train.loc[dfa_train['Mid size'] == 1.0, ['Small size', 'Large size']] = 0.0
dfa_train.loc[dfa_train['Small size'] == 1.0, ['Mid size', 'Large size']] = 0.0
dfa_train.loc[dfa_train['Large size'] == 1.0, ['Small size', 'Mid size']] = 0.0
dfa_train.loc[(dfa_train['Large size']==0.0)&(dfa_train['Mid size']==0.0),['Small size']] = 1.0
dfa_train.loc[(dfa_train['Mid size']==0.0)&(dfa_train['Small size']==0.0),['Large size']] = 1.0
dfa_train.loc[(dfa_train['Large size']==0.0)&(dfa_train['Small size']==0.0),['Mid size']] = 1.0
dfa_train.loc[(dfa_train['Large size'].isnull())&(dfa_train['Mid size'].isnull())&(dfa_train['Small size']==0),['Large size', 'Mid size']] = 0.5
dfa_train.loc[(dfa_train['Large size'].isnull())&(dfa_train['Mid size']==0)&(dfa_train['Small size'].isnull()),['Large size', 'Small size']] = 0.5
dfa_train.loc[(dfa_train['Large size']==0)&(dfa_train['Mid size'].isnull())&(dfa_train['Small size'].isnull()),['Small size', 'Mid size']] = 0.5
dfa_train.loc[(dfa_train['Large size'].isnull())&(dfa_train['Mid size'].isnull())&(dfa_train['Small size'].isnull()),['Small size', 'Mid size','Large size']] = 0.33

dfa_train['Mid size'] = dfa_train['Mid size']*2
dfa_train['Large size'] = dfa_train['Large size']*3
dfa_train['Size'] = dfa_train[['Small size','Mid size','Large size']].sum(axis=1)
dfa_train.drop(['Large size', 'Mid size', 'Small size'], axis=1, inplace=True)

dfa_train['NaN S'] = dfa_train[[ 'Small']].isnull().sum(axis=1)
dfa_train['NaN M'] = dfa_train[['Mid']].isnull().sum(axis=1)
dfa_train['NaN L'] = dfa_train[['Large']].isnull().sum(axis=1)
dfa_train.loc[dfa_train['Mid'] == 1.0, ['Small', 'Large']] = 0.0
dfa_train.loc[dfa_train['Small'] == 1.0, ['Mid', 'Large']] = 0.0
dfa_train.loc[dfa_train['Large'] == 1.0, ['Small', 'Mid']] = 0.0
dfa_train.loc[(dfa_train['Large']==0.0)&(dfa_train['Mid']==0.0),['Small']] = 1.0
dfa_train.loc[(dfa_train['Mid']==0.0)&(dfa_train['Small']==0.0),['Large']] = 1.0
dfa_train.loc[(dfa_train['Large']==0.0)&(dfa_train['Small']==0.0),['Mid']] = 1.0
dfa_train.loc[(dfa_train['Large'].isnull())&(dfa_train['Mid'].isnull())&(dfa_train['Small']==0),['Large', 'Mid']] = 0.5
dfa_train.loc[(dfa_train['Large'].isnull())&(dfa_train['Mid']==0)&(dfa_train['Small'].isnull()),['Large', 'Small']] = 0.5
dfa_train.loc[(dfa_train['Large']==0)&(dfa_train['Mid'].isnull())&(dfa_train['Small'].isnull()),['Small', 'Mid']] = 0.5
dfa_train.loc[(dfa_train['Large'].isnull())&(dfa_train['Mid'].isnull())&(dfa_train['Small'].isnull()),['Small', 'Mid','Large']] = 0.33

dfa_train['Mid'] = dfa_train['Mid']*2
dfa_train['Large'] = dfa_train['Large']*3
dfa_train['SML'] = dfa_train[['Small','Mid','Large']].sum(axis=1)
dfa_train.drop(['Large', 'Mid', 'Small'], axis=1, inplace=True)

dfa_train['Nan lession'] = dfa_train['Lession'].isnull().astype(int)
# Replace NaN values in 'Lession' with the mean of the column
mean_lession = dfa_train['Lession'].mean()
dfa_train['Lession'].fillna(mean_lession, inplace=True)

dfa_train['Nan GP'] = dfa_train['Genetic Propensity'].isnull().astype(int)
# Replace NaN values in 'Lession' with the mean of the column
mean_lession = dfa_train['Genetic Propensity'].mean()
dfa_train['Genetic Propensity'].fillna(mean_lession, inplace=True)

dfa_train['Nan ST'] = dfa_train['Skin X test'].isnull().astype(int)
# Replace NaN values in 'Lession' with the mean of the column
mean_lession = dfa_train['Skin X test'].mean()
dfa_train['Skin X test'].fillna(mean_lession, inplace=True)

dfa_train['Nan SC'] = dfa_train['Skin color'].isnull().astype(int)
# Replace NaN values in 'Lession' with the mean of the column
mean_lession = dfa_train['Skin color'].mean()
dfa_train['Skin color'].fillna(mean_lession, inplace=True)
trainAx = dfa_train.values
trainAy = dfa_trainres.values
```

```python
from sklearn import tree

clx = tree.DecisionTreeClassifier()
clx.fit(trainAx[:-100,:],trainAy[:-100])
```

```python
from sklearn import metrics

yhat = clx.predict(trainAx[-100:,:])
metrics.accuracy_score(trainAy[-100:],yhat[-100:])
```
