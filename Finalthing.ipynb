```python

def Aparam(dfa_train):  
    #dfa_train['NaN Size'] = dfa_train[['Mid size', 'Small size', 'Large size']].isnull().sum(axis=1)
    dfa_train['NaN SSize'] = dfa_train[[ 'Small size']].isnull().sum(axis=1)
    dfa_train['NaN MSize'] = dfa_train[['Mid size']].isnull().sum(axis=1)
    dfa_train['NaN LSize'] = dfa_train[['Large size']].isnull().sum(axis=1)
    dfa_train.loc[dfa_train['Mid size'] == 1.0, ['Small size', 'Large size']] = 0.0
    dfa_train.loc[dfa_train['Small size'] == 1.0, ['Mid size', 'Large size']] = 0.0
    dfa_train.loc[dfa_train['Large size'] == 1.0, ['Small size', 'Mid size']] = 0.0
    dfa_train.loc[(dfa_train['Large size']==0.0)&(dfa_train['Mid size']==0.0),['Small size']] = 1.0
    dfa_train.loc[(dfa_train['Mid size']==0.0)&(dfa_train['Small size']==0.0),['Large size']] = 1.0
    dfa_train.loc[(dfa_train['Large size']==0.0)&(dfa_train['Small size']==0.0),['Mid size']] = 1.0
    dfa_train.loc[(dfa_train['Large size'].isnull())&(dfa_train['Mid size'].isnull())&(dfa_train['Small size']==0),['Large size', 'Mid size']] = 0.5
    dfa_train.loc[(dfa_train['Large size'].isnull())&(dfa_train['Mid size']==0)&(dfa_train['Small size'].isnull()),['Large size', 'Small size']] = 0.5
    dfa_train.loc[(dfa_train['Large size']==0)&(dfa_train['Mid size'].isnull())&(dfa_train['Small size'].isnull()),['Small size', 'Mid size']] = 0.5
    dfa_train.loc[(dfa_train['Large size'].isnull())&(dfa_train['Mid size'].isnull())&(dfa_train['Small size'].isnull()),['Small size', 'Mid size','Large size']] = 0.33
    
    dfa_train['Mid size'] = dfa_train['Mid size']*2
    dfa_train['Large size'] = dfa_train['Large size']*3
    dfa_train['Size'] = dfa_train[['Small size','Mid size','Large size']].sum(axis=1)
    dfa_train.drop(['Large size', 'Mid size', 'Small size'], axis=1, inplace=True)
    
    dfa_train['NaN S'] = dfa_train[[ 'Small']].isnull().sum(axis=1)
    dfa_train['NaN M'] = dfa_train[['Mid']].isnull().sum(axis=1)
    dfa_train['NaN L'] = dfa_train[['Large']].isnull().sum(axis=1)
    dfa_train.loc[dfa_train['Mid'] == 1.0, ['Small', 'Large']] = 0.0
    dfa_train.loc[dfa_train['Small'] == 1.0, ['Mid', 'Large']] = 0.0
    dfa_train.loc[dfa_train['Large'] == 1.0, ['Small', 'Mid']] = 0.0
    dfa_train.loc[(dfa_train['Large']==0.0)&(dfa_train['Mid']==0.0),['Small']] = 1.0
    dfa_train.loc[(dfa_train['Mid']==0.0)&(dfa_train['Small']==0.0),['Large']] = 1.0
    dfa_train.loc[(dfa_train['Large']==0.0)&(dfa_train['Small']==0.0),['Mid']] = 1.0
    dfa_train.loc[(dfa_train['Large'].isnull())&(dfa_train['Mid'].isnull())&(dfa_train['Small']==0),['Large', 'Mid']] = 0.5
    dfa_train.loc[(dfa_train['Large'].isnull())&(dfa_train['Mid']==0)&(dfa_train['Small'].isnull()),['Large', 'Small']] = 0.5
    dfa_train.loc[(dfa_train['Large']==0)&(dfa_train['Mid'].isnull())&(dfa_train['Small'].isnull()),['Small', 'Mid']] = 0.5
    dfa_train.loc[(dfa_train['Large'].isnull())&(dfa_train['Mid'].isnull())&(dfa_train['Small'].isnull()),['Small', 'Mid','Large']] = 0.33
    
    dfa_train['Mid'] = dfa_train['Mid']*2
    dfa_train['Large'] = dfa_train['Large']*3
    dfa_train['SML'] = dfa_train[['Small','Mid','Large']].sum(axis=1)
    dfa_train.drop(['Large', 'Mid', 'Small'], axis=1, inplace=True)
    
    dfa_train['Nan lession'] = dfa_train['Lession'].isnull().astype(int)
    # Replace NaN values in 'Lession' with the mean of the column
    mean_lession = dfa_train['Lession'].mean()
    dfa_train['Lession'].fillna(mean_lession, inplace=True)
    
    dfa_train['Nan GP'] = dfa_train['Genetic Propensity'].isnull().astype(int)
    # Replace NaN values in 'Lession' with the mean of the column
    mean_lession = dfa_train['Genetic Propensity'].mean()
    dfa_train['Genetic Propensity'].fillna(-1.8, inplace=True)
    
    dfa_train['Nan ST'] = dfa_train['Skin X test'].isnull().astype(int)
    # Replace NaN values in 'Lession' with the mean of the column
    mean_lession = dfa_train['Skin X test'].mean()
    dfa_train['Skin X test'].fillna(3, inplace=True)
    
    dfa_train['Nan SC'] = dfa_train['Skin color'].isnull().astype(int)
    # Replace NaN values in 'Lession' with the mean of the column
    mean_lession = dfa_train['Skin color'].mean()
    dfa_train['Skin color'].fillna(0.5, inplace=True)
    dfa_train.loc[dfa_train['Doughnuts consumption'] > 1, 'Doughnuts consumption'] -= 1
    dfa_train = dfa_train.drop('Id', axis=1)
    return dfa_train
```

```python
import spacy
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

df = pd.read_csv('train_B_text.csv')
#Drop the id column
df = df.drop('Id', axis=1)
df.groupby('Fake/Real').describe()
df = df.drop_duplicates()
df.loc[:, 'Fake/Real'] = df['Fake/Real'].map({'fake': 0, 'real': 1})

Y = df['Fake/Real']
Y = Y.values
df = df.drop('Fake/Real', axis=1)
Y = np.array(Y,dtype=np.int64)

df['word_count'] = df['Title'].apply(lambda x: len(str(x).split()))
df['avg_word_length'] = df['Title'].apply(lambda x: np.mean([len(word) for word in str(x).split()]))
from textblob import TextBlob
df['polarity'] = df['Title'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['subjectivity'] = df['Title'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

additional_features = df[['word_count', 'avg_word_length','polarity','subjectivity']].values
additional_features = np.array(additional_features)

# Handle missing values (if any)
df = df.dropna(subset=['word_count', 'avg_word_length', 'polarity', 'subjectivity'])

# Now extract the values
additional_features = df[['word_count', 'avg_word_length', 'polarity', 'subjectivity']].values


titles =df['Title'].values

# Load spaCy's English model
nlp = spacy.load('en_core_web_sm')


# Function to process the titles and return POS tags
# Function to simplify POS tags to basic types
def simplify_pos(pos):
    pos_mapping = {
        'CCONJ': 'CONJ',  # Coordinating conjunction
        'SCONJ': 'CONJ',  # Subordinating conjunction
        'DET': 'DET',     # Determiner
        'NOUN': 'NOUN',   # Noun
        'PROPN': 'PROPN',  # Proper noun
        'ADJ': 'ADJ',     # Adjective
        'ADV': 'ADV',     # Adverb
        'VERB': 'VERB',   # Verb
        'ADP': 'ADP',     # Adposition (prepositions and postpositions)
        'AUX': 'VERB',    # Auxiliary verb
        'PART': 'PART',   # Particle
        'PRON': 'PRON',   # Pronoun
        'NUM': 'NUM',     # Numeral
        'PUNCT': 'PUNCT', # Punctuation
        'X': 'X'          # Other (foreign words, etc.)
    }
    return pos_mapping.get(pos, 'X')  # Default to 'X' for unknown tags

# Function to process the titles and return simplified POS tags
def pos_tags(text):
    doc = nlp(text)
    # Simplify POS tags for each word, removing punctuation and mapping to basic types
    pos_tags = [simplify_pos(token.pos_) for token in doc if token.pos_ != 'PUNCT']  # Remove punctuation
    return " ".join(pos_tags)

# Apply POS tagging to all titles
pos_titles = [pos_tags(title) for title in titles]

# Initialize the CountVectorizer to extract n-grams of simplified POS tags (1-2 grams)
vectorizer = CountVectorizer( stop_words='english', min_df=1,ngram_range=(1, 2), strip_accents='unicode',)
# Fit and transform the POS-tagged titles into a matrix of n-grams
X = vectorizer.fit_transform(pos_titles)

# Convert the matrix into a DataFrame to view the n-grams
ngram_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Remove columns where all values are zero
ngram_df_cleaned = ngram_df

# Optional: Remove columns with certain non-relevant n-grams, like "space"
ngram_df_cleaned = ngram_df_cleaned.loc[:, ~ngram_df_cleaned.columns.str.contains('space')]





from scipy.sparse import csr_matrix, hstack
# Assuming ngram_df_cleaned is a sparse matrix
additional_features_sparse = csr_matrix(additional_features)


# Combine ngram_df_cleaned and additional_features_sparse horizontally
X_combined = hstack([ngram_df_cleaned, additional_features_sparse])

X_train = X_combined
```
